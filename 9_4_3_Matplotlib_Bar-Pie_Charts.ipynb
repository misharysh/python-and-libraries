```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
np.random.seed(123)

groups = [i for i in range(7)]
counts = np.random.randint(3,10, len(groups))

plt.bar(groups,counts)
```

```python
np.random.seed(123)

groups = [i for i in range(7)]
counts = np.random.randint(3,10, len(groups))

plt.barh(groups,counts)
```

```python
import matplotlib.colors as mcolors

bc = mcolors.BASE_COLORS

np.random.seed(123)

groups = [i for i in range(7)]
counts = np.random.randint(0, len(bc), len(groups))

width = counts*0.1
colors = [['r','g','b'][int(np.random.randint(0,3,1))] for _ in counts]

plt.bar(groups,counts, width=width, alpha=0.5, bottom = 2,
       color=colors, edgecolor='k',linewidth=2)
```

```python
cat_par = [i for i in range(5)]

g1 = [10, 21, 34 ,12 ,27]
g2 = [17, 15, 25, 21, 26]

width = 0.3

x = np.arange(len(cat_par))

fix, ax = plt.subplots()

rects1 = ax.bar(x - width/2, g1, width, label='g1')
rects2 = ax.bar(x + width/2, g2, width, label='g2')

ax.set_title('Example of group diagram')
ax.set_xticks(x)
ax.set_xticklabels(cat_par)
ax.legend()
```

```python
np.random.seed(123)

groups = [i for i in range(7)]
counts = np.random.randint(3,10, len(groups))
error = 1

plt.bar(groups,counts, yerr=error, ecolor='r', alpha=0.5,
       edgecolor='b', linewidth=2)
```

```python
vals = [24, 17, 53, 21, 35]
labels = ['Ford', 'Toyota', 'BMW', 'Audi', 'Jaguar']

fig, ax = plt.subplots()
ax.pie(vals,labels=labels)
ax.axis('equal')
```

```python
vals = [24, 17, 53, 21, 35]
labels = ['Ford', 'Toyota', 'BMW', 'Audi', 'Jaguar']

explode = (0,0,0.3,0,0)

fig, ax = plt.subplots()
ax.pie(vals,labels=labels,autopct='%1.1f%%', shadow=True,
      explode=explode, 
       wedgeprops={'lw':1, 'ls': '--', 'edgecolor':'k'},
      rotatelabels=True)
ax.axis('equal')
```

```python
vals = [24, 17, 53, 21, 35]
labels = ["Ford", "Toyota", "BMW", "AUDI", "Jaguar"]
fig, ax = plt.subplots()
ax.pie(vals, labels=labels, wedgeprops=dict(width=0.5))
ax.axis('equal')
```
