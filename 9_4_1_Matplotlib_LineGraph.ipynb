```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [1, 5, 10, 15, 20]
y1 = [1, 7, 3, 5, 11]
y2 = [4, 3, 1, 8, 12]

plt.figure(figsize=(12,7))

plt.plot(x, y1, 'o-r', alpha=0.7, label='first', lw=5, mec='b',
        mew=2, ms=10)
plt.plot(x, y2, 'v-.g', label='second', mec='r', lw=2, mew=2,
        ms=12)

plt.legend()
plt.grid(True)
```

```python
import numpy as np
x = np.arange(0.0, 5, 0.01)
y = np.cos(x*np.pi)

plt.plot(x,y,c='r')
plt.fill_between(x,y)
```

```python
import numpy as np
x = np.arange(0.0, 5, 0.01)
y = np.cos(x*np.pi)

plt.plot(x,y,c='r')
plt.fill_between(x,y, where = (y > 0.75) | (y < -0.75))
```

```python
import numpy as np
x = np.arange(0.0, 5, 0.01)
y = np.cos(x*np.pi)


plt.plot(x,y,c='r')

plt.grid()

plt.fill_between(x,y, where=y>=0, color='g', alpha=0.3)
plt.fill_between(x,y, where=y<=0, color='r', alpha=0.3)
```

```python
x = np.arange(0.0, 5, 0.01)
y = np.cos(x*np.pi)

m_ev_case = [None, 10, (100,30), slice(100,400,15), [0,100, 200,300],
            [10,50,100]]

fig, ax = plt.subplots(2,3, figsize = (10, 7))
axs = [ax[i, j] for i in range(2) for j in range(3)]

for i, case in enumerate(m_ev_case):
    axs[i].set_title(str(case))
    axs[i].plot(x, y, 'o', ls='-', ms=7, markevery=case)
```
