```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.title('Title',alpha=0.2, color='g', fontsize=18, fontstyle='italic',
         fontweight='bold',linespacing=10)
plt.plot(range(0,10),range(0,10))
```

```python
from matplotlib.font_manager import FontProperties

plt.title('Title',
         fontproperties=FontProperties(family='monospace',
        style='italic',weight='heavy',size=15))
plt.plot(range(0,10),range(0,10))
```

```python
plt.title('Title',fontsize=17, position=(0.7,0.2),
         rotation='vertical')

plt.plot(range(0,10),range(0,10))
```

```python
from matplotlib.patches import FancyBboxPatch

bbox_properties=dict(
    boxstyle='rarrow,pad=0.3',
    ec='g',
    fc='r',
    ls='-',
    lw=3
)

plt.title('Title', fontsize=17, bbox=bbox_properties,
         position=(0.5,0.85))

plt.plot(range(0,10),range(0,10))
```
