```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [1,5,10,15,20]
y1 = [1, 7, 3, 5, 11]
y2 = [4, 3, 1, 8, 12]

plt.plot(x, y1, 'o-r', label='line 1')
plt.plot(x, y2, 'o-.g', label='line 2')
plt.legend()
```

```python
plt.plot(x, y1, 'o-b')
plt.plot(x, y2, 'o-.m')
plt.legend(['L1','L2'])
```

```python
locs = ['best', 'upper right', 'upper left', 'lower left', 
        'lower right', 'right', 'center left', 'center right', 
        'lower center', 'upper center', 'center']

plt.figure(figsize=(12,12))

for i in range(3):
    for j in range(4):
        if i*4 + j < 11:
            plt.subplot(3,4, i*4 + j + 1)
            plt.title(locs[i*4+j])
            plt.plot(x, y1, 'o-r', label='line 1')
            plt.plot(x, y2, 'o-.g', label='line 2')
            plt.legend(loc=locs[i*4+j])
            
        else:
            break
```

```python
plt.plot(x, y1, 'o-r',label='line 1')
plt.plot(x, y2, 'o-.g',label='line 2')
plt.legend(fontsize = 14, shadow=True, framealpha=1, facecolor='y',edgecolor='r', title = 'Legend')
```
