```python
import numpy as np
print 'x - array'
x = np.linspace(0,10,50)
y = x

plt.title('Linear dependence of y on x')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.plot(x,y,'r--')
```

```python
x = np.linspace(0,10,50);

y1 = x
y2 = [i**2 for i in x]

plt.title('Linear dependence of y1 on x and y2 on x^2')
plt.xlabel('x')
plt.ylabel('y1, y2')
plt.grid()
plt.plot(x,y1,x,y2)
```

```python
x = np.linspace(0,10,50);

y1 = x
y2 = [i**2 for i in x]

plt.figure(figsize=(9,9))

plt.subplot(2,1,1)
plt.plot(x,y1)
plt.title('Linear dependence of y1 on x and y2 on x^2')
plt.ylabel('y1')
plt.grid()

plt.subplot(2,1,2)
plt.plot(x,y2)
plt.xlabel('x')
plt.ylabel('y2')
plt.grid()
```
